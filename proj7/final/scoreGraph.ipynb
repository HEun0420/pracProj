```python
import matplotlib.pyplot as plt
import numpy as np

# 평가 지표와 값
metrics = ["BERTScore (P)", "BERTScore (R)", "BERTScore (F1)", "COMET Score"]
llama_scores = [0.6014, 0.6649, 0.6303, 0.5364]  # 모델 B (LLama 3.2-1B)
gorani_scores = [0.8977, 0.8955, 0.8965, 0.8275]  # 모델 A (Gorani)

# X축 위치
x = np.arange(len(metrics))
width = 0.35  # 바 너비

# 그래프 그리기
fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, llama_scores, width, label="LLama 3.2-1B", color="#dd8452")  # LLama 먼저
bars2 = ax.bar(x + width/2, gorani_scores, width, label="Gorani", color="#4c72b0")  # Gorani 나중

# 레이블 추가
ax.set_ylabel("Score")
ax.set_title("번역 평가 지표 비교")
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# 값 표시 (반올림 없이 그대로 출력)
for bars, scores in zip([bars1, bars2], [llama_scores, gorani_scores]):
    for bar, score in zip(bars, scores):
        height = bar.get_height()
        ax.annotate(f"{score}", xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=10, fontweight='bold')

# 출력
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
